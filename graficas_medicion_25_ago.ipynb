## **MEDICIONES 24 DE AGOSTO**

A continuación, se muestran las imágenes obtenidas para la intensidad de la luz que pasa por la botella y llega al soporte receptor del montaje experimental. Para ello, se mide la longitud de onda de la luz que pasa o que queda absorbida en la botella, dependiendo del color de las mismas.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# MEDICIÓN SIN BOTELLAS

```python
data_L_base = pd.read_csv("base_L.csv") 
data_i_base = pd.read_csv("base_i.csv") 
```

```python
fig = plt.figure(figsize=(5,5))
plt.plot(data_L_base,data_i_base, c='r')
plt.title('Medición sin presencia de Botellas')
plt.xlabel('Longitud de Onda (nm)')
plt.ylabel('Intensidad (Conteos)')

plt.grid()
plt.show()
```

# MEDICIÓN CON BOTELLA TRANSPARENTE

```python
data_L_t = pd.read_csv("transparente_L.csv") 
data_i_t = pd.read_csv("transparente_i.csv") 
```

```python
fig = plt.figure(figsize=(5,5))
plt.plot(data_L_t,data_i_t, c='b')
plt.title('Medición para Botella Transparente')
plt.xlabel('Longitud de Onda (nm)')
plt.ylabel('Intensidad (Conteos)')

plt.ylim(0,12000)
plt.grid()
plt.show()
```
